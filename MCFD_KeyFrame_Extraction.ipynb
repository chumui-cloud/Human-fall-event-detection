```python
import cv2
import os
import sys
sys.path.append('/content/drive/MyDrive/Colab Notebooks/')
import keyframe
import re
```

```python

n_key = 20
dir = '/content/drive/MyDrive/mcfd_processed_data'
save_dir = 'drive/MyDrive/mcfd_'+str(n_key)+'_frame/'
if not os.path.isdir(save_dir):
  os.mkdir(save_dir)
if not os.path.isdir(os.path.join(save_dir,'fall')):
  os.mkdir(os.path.join(save_dir,'fall'))
if not os.path.isdir(os.path.join(save_dir,'non_fall')):
  os.mkdir(os.path.join(save_dir,'non_fall'))
folders = ['fall','non_fall']
for folder in folders:
  folder_path = os.path.join(dir, folder)
  videos = os.listdir(folder_path)
  for video_name in videos:
    video = os.path.join(folder_path, video_name)
    tamp_val= int(re.search(r'\d+', video_name).group())
    if 'non_fall' not in video or (tamp_val>=105 and tamp_val<=1089):
      print(video +' done')
      continue
    try:
      kfs,labels = keyframe.get_keyframe_list_thread(video,n_key)
    except:
      continue
    cap = cv2.VideoCapture(video)
    frames = []
    for i in kfs:
      cap.set(1,i)
      _, frame = cap.read()
      frames.append(frame)
      if frame is None:
        continue
    size = frame.shape[1],frame.shape[0]
    fps = cap.get(cv2.CAP_PROP_FPS)
    if 'fall' in video:
      path  = os.path.join(save_dir, 'fall/'+video_name)
    if 'non_fall' in video:
      path = os.path.join(save_dir, 'non_fall/'+video_name)
    out = cv2.VideoWriter(path,cv2.VideoWriter_fourcc(*'mp4v'), fps, size)
    for f in range(len(frames)):
      out.write(frames[f])
    out.release()
    print(video +' done')
  
```
