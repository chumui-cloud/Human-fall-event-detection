```python
import os
import cv2
import numpy as np
import random
import sys
sys.path.append('/content/drive/MyDrive/Colab Notebooks')
```

```python
def process_data(dataset_name, extracted_frames, size):
  if dataset_name == 'mcfd':
    dir = 'drive/MyDrive/mcfd_'+str(extracted_frames)+'_segmented/'
  elif dataset_name == 'urfd':
    dir = 'drive/MyDrive/urfd_'+str(extracted_frames)+'_frame/'
  folders = ['fall','non_fall']
  clips= []
  Y=[]
  for folder in folders:
    folder_path = os.path.join(dir, folder)
    videos = os.listdir(folder_path)
    for video in videos:
      video = os.path.join(folder_path, video)
      count = 0
      cap = cv2.VideoCapture(video)
      if 'non_fall' in video:
        Y.append(0)
      else:
        Y.append(1)
      list_item_frames = []
      print(video)
      while count< extracted_frames:
        success, frame = cap.read()
        count = count+1     
        frame = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        new_frame = cv2.resize(frame, size)
        list_item_frames.append(new_frame)
      clips.append(list_item_frames)
  X = np.array(clips)
  Y = np.array(Y)
  return (X, Y)
```

```python
data, label = process_data('urfd', 10, (224,224))
```

```python
import matplotlib.pyplot as plt
def plot_frames(video_path):
  cap = cv2.VideoCapture(video_path)
  n_frame = int(cap. get(cv2. CAP_PROP_FRAME_COUNT))
  print(n_frame)
  start = 3
  for x in range(start,1000):
    if n_frame%x==0:
      y = int(n_frame/x)
      break
  f, ax = plt.subplots(x,y,figsize=(12,12))
  i = 0
  for ax1 in ax:
    for ax2 in ax1:
      cap.set(1,i);
      success, frame = cap.read()
      i = i+1
      ax2.imshow(cv2.cvtColor(frame,cv2.COLOR_BGR2RGB))
      ax2.axis('off')
plot_frames('drive/MyDrive/urfd_20_segmented/non_fall/urfd_non_fall1.avi')
```
